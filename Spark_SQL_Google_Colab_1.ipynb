```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("name").getOrCreate()
```

```python
w = spark.read.format('csv')\
.option('header','True').option('inferschema','True').option('mode','failfast').load('/content/Worker.csv')
```

```python
p = spark.read.format('csv')\
.option('header','True').option('inferschema','True').option('mode','failfast').load('/content/Projects.csv')
```

```python
p.createOrReplaceTempView("projects")
```

```python
spark.sql('select * from projects').show()
```

```python
w.createOrReplaceTempView('worker')
p.createOrReplaceTempView('projects')
```

```python
spark.sql('select * from worker').show()
```

```python
w.select('*').show()
```

```python
spark.sql('select first_name, last_name from worker').show()
```

```python
w.select('first_name','last_name').show()
```

```python
#where
spark.sql('select * from worker where salary > 15000').show()
w.where('salary>15000').show()
```

```python
from pyspark.sql.functions import col
w.where(col('SALARY')>15000).show()
```

```python
w.where(w.SALARY>15000).show()
```

```python
#and ,or
#hr dept and salary>15000

spark.sql("SELECT * FROM worker WHERE department='HR' AND salary > 15000").show()
w.where((col('DEPARTMENT')=='HR') & (col('SALARY')>15000)).show()
```

```python
from pyspark.sql.functions import *
```

```python
#HR DEPT ~= 15000

w.where((col('DEPARTMENT')=="HR") &~(col('SALARY')>15000)).show()
```

```python
w.where(~(col('DEPARTMENT')=="HR") &~(col('SALARY')==15000)).show()
```

```python
#whose first_name start with A
w.where('FIRST_NAME like "A%"').show()
w.where(w.FIRST_NAME.startswith("A")).show()
w.where(col('FIRST_NAME').startswith("A")).show()
```

```python
#endswith, contain
#first name end with A

w.where('FIRST_NAME like "%a"').show()
w.where(w.FIRST_NAME.endswith("a")).show()
w.where(col('FIRST_NAME').endswith("a")).show()
```

```python
#group by
#agg

w.groupBy('DEPARTMENT').agg(sum('SALARY').alias('total_salary')).show()
```

```python
#orderby
w.groupBy('DEPARTMENT').agg(sum('SALARY').alias('total_salary')).orderBy(col('total_salary').desc()).show()
```

```python
#department whose total >
w.groupBy('DEPARTMENT') \
    .agg(sum('SALARY').alias('total_salary')) \
    .where(col('total_salary') > 900000) \
    .orderBy(col('total_salary').desc()) \
    .show()
```

```python
#worker belong to HR or admin
w.where((w.DEPARTMENT == 'HR') | (w.DEPARTMENT == 'ADMIN')).show()
```

```python
w.groupBy('DEPARTMENT').agg(sum('SALARY').alias('Total'))\
.where(col('Total')>200000).show()
```

```python
#order by
w.orderBy(col('SALARY').desc()).show()
w.orderBy(desc('SALARY')).show()
```

```python
#top 2 salary
w.orderBy(col('SALARY').desc()).limit(2).show()
w.orderBy(desc('SALARY')).limit(2).show()
```

```python
#new column -- new_salary where all salary should increase by 100
w.withColumn('new_salary',col('SALARY')+100).show()
```

```python
#joins -- inner join
result = w.join(p, w["WORKER_ID"] == p["WORKER_ID"], "inner")
result.show()
```

```python
#worker name and project name, they are working on and salary > 500000
w.join(p, w["WORKER_ID"] == p["WORKER_ID"], "inner")\
.where ('salary>50000')\
.select('FIRST_NAME','PROJECT_NAME').show()
```

```python
#Left Join
w.join(p,w["WORKER_ID"]==p["WORKER_ID"],"left").show()
```

```python
#leftanti -- remaining records from left table which are not common
w.join(p,w['WORKER_ID']==p['WORKER_ID'],'leftanti').show()
```

```python
w.join(p,w['WORKER_ID']==p['WORKER_ID'],'leftsemi').show()
```

```python
#right join
#all the worker and project name, if they have any else keep it null
p.join(w,w['WORKER_ID']==p['WORKER_ID'],'right').show()
w.join(p,w['WORKER_ID']==p['WORKER_ID'],'right').show()
```

```python
#full
w.join(p,w['WORKER_ID']==p['WORKER_ID'],'full').show()
```
