```python
# any = how
df_pyspark.na.drop(how='all').show() #it will be only applied when all 4 columns are null but here in case of Mahesh there is name and salary
```

```python
#any
df_pyspark.na.drop(how='any').show()
```

```python
#Threshold
df_pyspark.na.drop(how='any', thresh=2).show() #thresh = 2 means in a row atleast 2 non-null values should be there
```

```python
df_pyspark.na.drop(how='any', thresh=3).show() #thresh = 3 means in a row atleast 3 non-null values should be there
```

```python
df_pyspark.na.drop(how='any', subset=['Experience']).show() # Want to delete null values from Experience
```
